```python
import pandas as pd
import plotly.express as px
import json
```

```python
jsonl_path = '../data/processed/reports_texts_clean.jsonl'
```

```python
reports_list = []

with open(jsonl_path, 'r', encoding='utf-8') as jsonl_file:
    for row in jsonl_file:
        report = json.loads(row)
        reports_list.append(report)

reports_data = pd.DataFrame([
    {
        'nazwa_raportu': report['nazwa_raportu'],
        'dlugosc_tekstu': len(report['text']),
        **report['labels']
    }
    for report in reports_list
])
```

```python
print(f'Liczba raportów: {len(reports_data)}')
```

```python
fig = px.histogram(data_frame=reports_data,
                   x='dlugosc_tekstu',
                   labels={'dlugosc_tekstu': 'Długość tekstu (znaki)'},
                   nbins=30)

fig.update_layout(
    yaxis_title='Liczba raportów',
    xaxis_tickformat=',.0f',
    bargap=0.2)

fig.update_traces(
    texttemplate="%{y}",
    textposition="outside")

fig.show()
```

```python
# Wczytanie ocen ekspertów
scores = pd.read_csv('../data/expert_scores.csv', encoding='utf-8', sep=";")
```

```python
criteria_grades = {
    '1': [0, 1.0],
    '2': [0, 1.0],
    '3': [0, 0.25, 0.5, 0.75, 1.0],
    '4': [0, 1.0],
    '5': [0, 1.0],
    '6': [0, 1.0],
    '7': [0, 1.0],
    '8': [0, 1.0],
    '9': [0, 1.0],
}

for criteria, grades in criteria_grades.items():
    values = scores[criteria].value_counts().sort_index()
    print(values)
    fig = px.bar(
        x=values.index,
        y=values.values,
        title=f'Rozkład ocen dla kryterium {criteria}')
    
    fig.update_layout(
        xaxis_type='category',
        yaxis_title='Liczba raportów',
        xaxis_title='Oceny')
    
    fig.update_traces(
        texttemplate="%{y}",
        textposition="outside")

    fig.show()
```

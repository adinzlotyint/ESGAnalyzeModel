```python
import pandas as pd
import plotly.express as px
```

```python
# Wczytanie ocen ekspertów
oceny_ekspertow = pd.read_csv('../data/processed/expert_scores_uproszczone_po_konwersji.csv', encoding='utf-8')
```

```python
kryteria_przedzialy = {
    'C1': [0, 0.5, 1.0],
    'C2': [0, 0.33, 0.67, 1.0],
    'C3': [0, 0.5, 1.0],
    'C4': [0, 0.25, 0.5, 0.75, 1.0],
    'C5': [0, 0.33, 0.67, 1.0],
    'C6': [0, 1.0],
    'C7': [0, 1.0],
    'C8': [0, 0.5, 1.0],
    'C9': [0, 1.0],
    'C10': [0, 0.33, 0.5, 0.67, 1.0],
    'C11': [0, 1.0],
    'C12': [0, 1.0],
}

for kryteria, przedzialy in kryteria_przedzialy.items():
    wartosci = oceny_ekspertow[kryteria].value_counts().sort_index()
    print(wartosci)
    fig = px.bar(
        x=wartosci.index,
        y=wartosci.values,
        title=f'Rozkład ocen dla kryterium {kryteria}')
    
    fig.update_layout(
        xaxis_type='category',
        yaxis_title='Liczba raportów',
        xaxis_title='Ocena')
    
    fig.update_traces(
        texttemplate="%{y}",
        textposition="outside")

    fig.show()
```

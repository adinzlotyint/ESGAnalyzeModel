```python
import pandas as pd
import plotly.express as px
import json
```

```python
jsonl_path = '../data/processed/teksty_raportow_wyczyszczone.jsonl'
```

```python
# Wczytanie raportów
lista_raportów = []

with open(jsonl_path, 'r', encoding='utf-8') as plik_jsonl:
    for wiersz in plik_jsonl:
        raport = json.loads(wiersz)
        lista_raportów.append(raport)

dane_raporty = pd.DataFrame([
    {
        'nazwa_raportu': raport['nazwa_raportu'],
        'dlugosc_tekstu': len(raport['text']),
        **raport['labels']
    }
    for raport in lista_raportów
])
```

```python
print(f'Liczba raportów: {len(dane_raporty)}')
```

```python
fig = px.histogram(data_frame=dane_raporty,
                   x='dlugosc_tekstu',
                   labels={'dlugosc_tekstu': 'Długość tekstu (znaki)'},
                   nbins=30)

fig.update_layout(
    yaxis_title='Liczba raportów',
    xaxis_tickformat=',.0f',
    bargap=0.2)

fig.update_traces(
    texttemplate="%{y}",
    textposition="outside")

fig.show()
```
